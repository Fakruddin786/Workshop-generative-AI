```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","April","May","Jun"]
energy_consumption=[1200,1300,1100,1500,1400,1000]

plt.plot(months, energy_consumption, marker='*', color='b', linestyle=':')
plt.title('Energy Consumption Over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (MWh)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["Jan","Feb","Mar","April","May","Jun"]
energy_consumption=[1200,1300,1100,1500,1400,1000]

plt.bar(months, energy_consumption, color='g')
plt.title('Energy Consumption Over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
#Pie chart to show energy consumption
plt.pie(energy_consumption, labels = months, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'coral'])
plt.title('Energy Consumption Share By Source')
#Add a title
plt.show()
```
